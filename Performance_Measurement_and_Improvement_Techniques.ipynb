# Performance Measurement and Improvement Techniques

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## Measuring Performance with OpenCV

```python
# image processing involves large number of operations per second. therefore along with the correctness of the solution its computaional speed is also important.
```

```python
# The cv.getTickCount function returns the number of clock-cycles after a reference event
# (like the moment the machine was switched ON) to the moment this function is called.
# So if you call it before and after the function execution, you get the number of clock-cycles used to execute a function.
```

```python
# lets find out time taken to simple read and plot an image

e1 = cv.getTickCount()
img = cv.imread('bird.jpg')
plt.imshow(img)
e2 = cv.getTickCount()
time = (e2 - e1)/ cv.getTickFrequency()
time
```

# Default Optimization in OpenCV

```python
# OpenCV runs the optimized code if it is enabled, otherwise it runs the unoptimized code.
# You can use cv.useOptimized() to check if it is enabled/disabled and cv.setUseOptimized() to enable/disable it.
# Let's see a simple example.
```

```python
cv.useOptimized()
```

```python
%timeit res = cv.medianBlur(img,49)
```

```python
cv.setUseOptimized(False)
cv.useOptimized()
```

```python
%timeit res = cv.medianBlur(img,49)
```
